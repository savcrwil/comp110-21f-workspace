## Selecting Subsets of Data from a DataFrame

```python
titanic = pd.read_csv("../data/titanic.csv")
```

```python
ages = titanic["Age"]

ages.head()
```

```python
type(titanic["Age"])
```

```python
titanic["Age"].shape
```

```python
age_sex = titanic[["Age", "Sex"]]

age_sex.head()
```

```python
type(titanic[["Age", "Sex"]])
```

```python
titanic[["Age", "Sex"]].shape
```

```python
above_35 = titanic[titanic["Age"] > 35]

above_35.head()
```

```python
titanic["Age"] > 35
```

```python
above_35.shape
```

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]

class_23.head()
```

```python
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]

class_23.head()
```

```python
age_no_na = titanic[titanic["Age"].notna()]

age_no_na.head()
```

```python
age_no_na.shape
```

```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]

adult_names.head()
```
